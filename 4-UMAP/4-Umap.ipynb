```python
def draw_group(x, labels, gidx, savedir=None):
    xg = x[labels==gidx]
    
    fig = plt.figure(figsize=(8,4))
    fig.suptitle(gidx, fontsize=20)

    for i in range(4):
        plt.subplot(1,4,i+1)
        plt.imshow(xg[i].reshape(128,64), cmap='gray')
        plt.xticks([])
        plt.yticks([])
    
    
    if savedir != None:
        plt.savefig(savedir+'/group_%i'%gidx)
    plt.show()
```
