```python
def process(datadir, resdir):
    ua = UmapAnalysis(datadir, resdir)
    umaped = ua.umap_embedding(ua.xlat)
    labels = ua.kmeans_clustering(umaped, 12)
    return ua, umaped, labels
```

```python
labels = ua.kmeans_clustering(umaped, 9)
```
